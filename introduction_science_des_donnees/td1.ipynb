```python
california = fetch_california_housing()
california_df = pd.DataFrame(california.data, columns=california.feature_names)
california_df.head()
```

# Q3-bis

```python
df_missing = california_df.copy()
df_missing
```

```python
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Introduce 2% of NaN values in AveRooms column
n_missing_averooms = int(0.02 * len(df_missing))
missing_indices_averooms = np.random.choice(df_missing.index, n_missing_averooms, replace=False)
df_missing.loc[missing_indices_averooms, 'AveRooms'] = np.nan

# Introduce 2% of NaN values in HouseAge column
n_missing_houseage = int(0.02 * len(df_missing))
missing_indices_houseage = np.random.choice(df_missing.index, n_missing_houseage, replace=False)
df_missing.loc[missing_indices_houseage, 'HouseAge'] = np.nan

# Display the result
print(f"Introduced {df_missing['AveRooms'].isnull().sum()} NaN values in AveRooms ({df_missing['AveRooms'].isnull().mean()*100:.2f}%)")
print(f"Introduced {df_missing['HouseAge'].isnull().sum()} NaN values in HouseAge ({df_missing['HouseAge'].isnull().mean()*100:.2f}%)")
df_missing.head(20)
```

# Q5 - Boxplots

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Add target variable to dataframe
california_df['MedHouseVal'] = california.target

# Create boxplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].boxplot(california_df['MedHouseVal'])
axes[0].set_title('MedHouseVal')
axes[0].set_ylabel('Value')

axes[1].boxplot(california_df['AveRooms'])
axes[1].set_title('AveRooms')
axes[1].set_ylabel('Value')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(california_df['MedHouseVal'], bins=50, edgecolor='black')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.title('Distribution of Median House Value')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
Q1 = california_df.quantile(0.25)
Q3 = california_df.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = {}
for col in california_df.columns:
    outlier_count = ((california_df[col] < lower_bound[col]) | (california_df[col] > upper_bound[col])).sum()
    outliers_iqr[col] = outlier_count
    print(f"{col}: {outlier_count} outliers")
```

```python
from scipy import stats

outliers_zscore = {}
for col in california_df.columns:
    z_scores = np.abs(stats.zscore(california_df[col]))
    outlier_count = (z_scores > 3).sum()
    outliers_zscore[col] = outlier_count
    print(f"{col}: {outlier_count} outliers")
```

```python
for col in california_df.columns:
    print(f"{col}: IQR={outliers_iqr[col]}, Z-Score={outliers_zscore[col]}")
```

```python
# PARTIE 4 - Analyse de Corrélation

import seaborn as sns
# Q9 - Corrélation - Calculer matrice de corrélation
correlation_matrix = california_df.corr()

# Identifier top 3 corrélations avec MedHouseVal
top_3_correlations = correlation_matrix['MedHouseVal'].sort_values(ascending=False)[1:4]
print("Top 3 corrélations avec MedHouseVal:")
print(top_3_correlations)

# Visualiser heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Matrice de Corrélation')
plt.tight_layout()
plt.show()
```
