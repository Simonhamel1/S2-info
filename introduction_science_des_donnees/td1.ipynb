## Q7

```python
Q1 = california_df.quantile(0.25)
Q3 = california_df.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = {}
for col in california_df.columns:
    outlier_count = ((california_df[col] < lower_bound[col]) | (california_df[col] > upper_bound[col])).sum()
    outliers_iqr[col] = outlier_count
    print(f"{col}: {outlier_count} outliers")
```

## Q7

```python
from scipy import stats

outliers_zscore = {}
for col in california_df.columns:
    z_scores = np.abs(stats.zscore(california_df[col]))
    outlier_count = (z_scores > 3).sum()
    outliers_zscore[col] = outlier_count
    print(f"{col}: {outlier_count} outliers")
```

## Q8


```python
for col in california_df.columns:
    print(f"{col}: IQR={outliers_iqr[col]}, Z-Score={outliers_zscore[col]}")
```

```python
from sklearn.preprocessing import StandardScaler

# Apply StandardScaler to california_df features
scaler_standard = StandardScaler()
california_df_standardized = pd.DataFrame(scaler_standard.fit_transform(california_df), columns=california_df.columns)

# Verify mean ≈ 0 and std ≈ 1
print("Mean values after standardization:")
print(california_df_standardized.mean())
print("\nStandard deviation after standardization:")
print(california_df_standardized.std())

# Display first few rows
california_df_standardized.head()
```

```python
# Q14 - Comparaison des distributions avant/après scaling

import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Original distribution
axes[0, 0].hist(california_df['MedInc'], bins=50, edgecolor='black')
axes[0, 0].set_title('Original - MedInc')
axes[0, 0].set_xlabel('Value')
axes[0, 0].set_ylabel('Frequency')

# MinMax scaled distribution
axes[0, 1].hist(california_df_scaled['MedInc'], bins=50, edgecolor='black', color='orange')
axes[0, 1].set_title('MinMax Scaled - MedInc')
axes[0, 1].set_xlabel('Value')
axes[0, 1].set_ylabel('Frequency')

# Standardized distribution
axes[1, 0].hist(california_df_standardized['MedInc'], bins=50, edgecolor='black', color='green')
axes[1, 0].set_title('Standardized - MedInc')
axes[1, 0].set_xlabel('Value')
axes[1, 0].set_ylabel('Frequency')

axes[1, 1].axis('off')
axes[1, 1].text(0.1, 0.5, 
    "Discussion:\n"
    "• MinMax scaling (0-1) compresse les données dans une plage fixe, préservant la distribution.\n"
    "• Standardization (z-score) centre les données (μ=0, σ=1) sans limiter la plage.\n"
    "• Les deux transformations préservent la forme de la distribution originale.\n"
    "• MinMax est préférable pour les algorithmes sensibles à l'échelle (KNN, SVM).",
    fontsize=10, verticalalignment='center', family='monospace')

plt.tight_layout()
plt.show()
```

```python
# Q15 - Equal-width binning for MedHouseVal into 5 equal intervals

# Get min and max values
min_val = california_df['MedHouseVal'].min()
max_val = california_df['MedHouseVal'].max()

# Create 5 equal-width bins
bins = pd.cut(california_df['MedHouseVal'], bins=5)

# Display bin information
print(f"Min: {min_val}, Max: {max_val}")
print(f"\nBin intervals:")
print(bins.value_counts().sort_index())

# Add binned column to dataframe
california_df['MedHouseVal_binned'] = bins

# Display first few rows
print("\nFirst 10 rows with binned values:")
print(california_df[['MedHouseVal', 'MedHouseVal_binned']].head(10))
```

```python
# Q16 - Equal-frequency binning for MedHouseVal into 5 equal-frequency intervals

# Create 5 equal-frequency bins using qcut
bins_qcut = pd.qcut(california_df['MedHouseVal'], q=5, duplicates='drop')

# Display bin information
print(f"Bin intervals (Equal-Frequency):")
print(bins_qcut.value_counts().sort_index())

# Add binned column to dataframe
california_df['MedHouseVal_binned_qcut'] = bins_qcut

# Display first few rows
print("\nFirst 10 rows with equal-frequency binned values:")
print(california_df[['MedHouseVal', 'MedHouseVal_binned_qcut']].head(10))
```

```python
# Q17 - Comparaison des impacts des outliers selon les méthodes de détection

print("Outlier Detection Comparison:")
print("=" * 60)
print(f"\nIQR Method detected {sum(outliers_iqr.values())} total outliers")
print(f"Z-Score Method detected {sum(outliers_zscore.values())} total outliers")

print("\n\nDetailed breakdown by feature:")
print(f"{'Feature':<15} {'IQR':<10} {'Z-Score':<10} {'Difference'}")
print("-" * 60)

# Only iterate over features that exist in both dictionaries
for col in outliers_iqr.keys():
    iqr_count = outliers_iqr[col]
    zscore_count = outliers_zscore[col]
    diff = iqr_count - zscore_count
    print(f"{col:<15} {iqr_count:<10} {zscore_count:<10} {diff:+d}")

print("\n\nInterpretation:")
print("• IQR method is more sensitive and detects more outliers overall")
print("• Z-Score requires extreme deviations (>3σ) and detects fewer outliers")
print("• Feature 'AveBedrms' shows largest difference (1424 vs 145)")
print("• Both methods agree on features with minimal outliers (HouseAge, Latitude, Longitude)")
```
