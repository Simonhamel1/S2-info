```python
import scipy.interpolate
import pandas as pd


tab = pd.DataFrame()
tab['i'] = [0, 1, 2, 3, 4]
tab['x'] = [0, 1, 2, 3, 5]
tab['f(x)'] = [0.002, 0.04, 0.1, 0.08, 0.01]

spline = scipy.interpolate.CubicSpline(tab['x'], tab['f(x)'], bc_type='natural')

import matplotlib.pyplot as plt
import numpy as np

plt.scatter(tab['x'], tab['f(x)'], color='red', label='Data Points')
x_new = np.linspace(min(tab['x']), max(tab['x']), 100)
y_new = spline(x_new)
plt.plot(x_new, y_new, color='blue', label='Cubic Spline Interpolation')

plt.show()
```
