```python
from math import *

def methode_fixe(fct, epsilon, Max):
    n = 0
    xn = 0 # a changer selon la fonction
    xn1 = fct(xn)
    while abs(xn1 - xn)/abs(xn1) > epsilon and n < Max:
        xn, xn1 = xn1, fct(xn1)
        n += 1
    return xn1, n
```

```python
fct1 = lambda x: sqrt(2 * x + 3)

solution, iterations = methode_fixe(fct1, 0.001, 100)
print("Solution approchée :", solution)
print("Nombre d'itérations :", iterations)
```

```python
fct1 = lambda x: 3 / (x - 2)
solution, iterations = methode_fixe(fct1, 0.00000001, 100)
print("Solution approchée :", solution)
print("Nombre d'itérations :", iterations)
```

```python
fct1 = lambda x: (x**2 - 3) / 2

solution, iterations = methode_fixe(fct1, 0.001, 1000)
print("Solution approchée :", solution)
print("Nombre d'itérations :", iterations)
```

# Exo 2

```python
def methode_newton(fct, dfct, epsilon, Max, x0):
    n = 0
    xn = x0 # a changer selon la fonction
    xn1 = xn - fct(xn) / dfct(xn)
    while abs(xn1 - xn)/abs(xn1) > epsilon and n < Max:
        xn = xn1
        xn1 = xn1 - fct(xn1) / dfct(xn1)
        n += 1
    return xn1, n
```

```python
fct = lambda x: x**2 - 2
dfct = lambda x: 2 * x

solution, iterations = methode_newton(fct, dfct, 0.001, 100, 1)
print("Solution approchée :", solution)
print("Nombre d'itérations :", iterations)
```

```python
# Derivative of t(x) for Newton's method (maximize thickness)
fct = lambda x: (3.7 / (2 * sqrt(x))) - 3.4 - 1.2 * x**3
dfct = lambda x: - (3.7 / (4 * x**(3/2))) - 3.6 * x**2

solution, iterations = methode_newton(fct, dfct, 0.0001, 100, 0.1)
print("Solution approchée :", solution)
print("Nombre d'itérations :", iterations)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def loi_exp(x, theta):
    return theta * np.exp(-theta * x)

def loi_uniforme(x, a, b):
    return np.where((x >= a) & (x <= b), 1 / (b - a), 0)

def loi_normale(x, mu, sigma):
    coeff = 1 / (sigma * np.sqrt(2 * np.pi))
    exponent = -0.5 * ((x - mu) / sigma) ** 2
    return coeff * np.exp(exponent)

if __name__ == "__main__":
    x = np.linspace(0, 5, 1000)

    plt.figure(figsize=(12, 8))

    # Loi exponentielle
    plt.subplot(3, 1, 1)
    plt.plot(x, loi_exp(x, theta=1), label='Loi Exponentielle (θ=1)', color='blue')
    plt.title('Loi Exponentielle')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid()

    # Loi uniforme
    plt.subplot(3, 1, 2)
    plt.plot(x, loi_uniforme(x, a=1, b=4), label='Loi Uniforme [1, 4]', color='green')
    plt.title('Loi Uniforme')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid()

    # Loi normale
    plt.subplot(3, 1, 3)
    plt.plot(x, loi_normale(x, mu=2.5, sigma=0.5), label='Loi Normale (μ=2.5, σ=0.5)', color='red')
    plt.title('Loi Normale')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()
```
