```python
import numpy as np
import matplotlib.pyplot as plt

#PARAMS:
N = 10000
theta = 0.5
a,b = 0,5
delta = 0.05

loi_exp = lambda x, theta: theta * np.exp(-theta * x)

def sim_exp(N,theta, a,b, delta):
    Ech = np.zeros(N)
    for i in range(N):
        u = np.random.rand()
        Ech[i] = -np.log(1-u)/theta
    
    return Ech

densite_emp = []

for k in range(int((b-a)/delta)):
    ai = a + k*delta
    bi = a + (k+1)*delta
    Ech = sim_exp(N, theta, a, b, delta)
    ni = np.sum((Ech >= ai) & (Ech < bi))
    densite_emp.append(ni / (N * delta))
    
#function theorique
plt.figure(figsize=(8,5))

# tableau des milieux d'intervalles
valeur = []
i = 0
while i < len(densite_emp):
    valeur.append(a + (i + 0.5) * delta)
    i += 1

plt.plot(valeur, densite_emp, 'bo', alpha=0.7, label="Densité empirique")
# Densité théorique
x = np.linspace(a, b, 100) # points pour tracer la courbe théorique
plt.plot(x, loi_exp(x, theta), 'r-', lw=2, label="Densité théorique")
plt.xlabel('x')
plt.ylabel('Densité')
plt.title("Comparaison densité empirique et théorique (loi exponentielle)")
plt.legend()
plt.show()
```
