```python
import numpy as np
from math import sqrt, log, cos, sin, pi

# Génération Box-Muller (loi normale)
def loi_normale(n):
    u1 = np.random.rand(n)
    u2 = np.random.rand(n)
    z0 = np.sqrt(-2.0 * np.log(u1)) * np.cos(2.0 * np.pi * u2)
    z1 = np.sqrt(-2.0 * np.log(u1)) * np.sin(2.0 * np.pi * u2)
    return z0, z1

def generer_echantillon_normale(n):
    z0, z1 = loi_normale(n)
    return np.concatenate((z0, z1))

# (optionnel) Intégrale sous la loi normale : E[g(Z)] = ∫ g(x)φ(x)dx
def integrale_sous_normale(g, n):
    z = generer_echantillon_normale(n)  # retourne n valeurs environ
    return np.mean(g(z))

# Exemple : ∫_0^1 x^2 dx = 1/3
approx = integrale_sous_normale(lambda x: (3*(2*pi)**0.5)/(1+x**2), n=1_0000_000)
print("Approximation ∫_0^1 x^2 dx =", approx)
```

```python
import scipy.integrate as integrate

def g(x):
    return (3)/(1+x**2)*np.exp(-x**2/2)

result, error = integrate.quad(g, -np.inf, np.inf)
print("Intégrale exacte ∫_0^1 x^2 dx =", result)
print("l'erreur est =", abs(result - approx))
```

## utilisation de la loi de poisson pour une approximation de la somme

```python
import numpy as np
from math import factorial

def approx_somme_mc(N=100_000, Kmax=40):
    # Indices
    k = np.arange(Kmax + 1)

    # Factorielle vectorisée proprement
    fact = np.array([factorial(int(ki)) for ki in k], dtype=float)

    # Terme de la somme
    termes = 3.0**k / (fact * np.sqrt(k**2 + 1.0))

    # Loi de tirage p_k ∝ 3^k / k!
    numerateur = 3.0**k / fact
    proba = numerateur / numerateur.sum()

    # Tirages selon p_k
    K = np.random.choice(k, size=N, p=proba)

    # Importance sampling : a_k / p_k
    vals = termes[K] / proba[K]

    return vals.mean()

S_hat = approx_somme_mc(N=1_000_000, Kmax=40)
print("Approximation Monte Carlo =", S_hat)
```

# somme de sqrt(k**3 +5)/3**k

```python
import numpy as np

def generer_geometrique(N, p):
    """Loi géométrique sur N* : P(X=k) = p*(1-p)^(k-1), k >= 1"""
    u = np.random.rand(N)
    return np.ceil(-np.log(u) / np.log(1 / (1 - p))).astype(int)
    # Equivalent : np.floor(np.log(u) / np.log(1-p)).astype(int) + 1

N = 1_000_000
p = 2/3

ech = generer_geometrique(N, p)  # X ~ G(2/3), valeurs dans {1, 2, 3, ...}

# S2 = E[g(X)] avec g(x) = (1/2) * sqrt(x^3 + 5)
g = lambda x: (1/2) * np.sqrt(x**3 + 5)

approx_S2 = np.mean(g(ech))

print(f"Approximation Monte Carlo de S2 = {approx_S2:.6f}")
```
