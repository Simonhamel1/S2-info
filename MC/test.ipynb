```python
import numpy as np
import pandas as pd

def simul_loi_uniforme():
    return np.random.random()

def simul_loi_bernoulli(p):
    return 1 if simul_loi_uniforme() < p else -1

chemin = []
def creation_tab(nmbre_chemin = 10, nombre_de_lancers = 10, probabilite = 0.5):
    for trace in range(nmbre_chemin):
        chemin.append([])
        for i in range(nombre_de_lancers):
            chemin[trace].append(chemin[trace][-1] + simul_loi_bernoulli(probabilite) if chemin[trace] else simul_loi_bernoulli(probabilite))

creation_tab(nmbre_chemin=10000, nombre_de_lancers=100, probabilite=0.5)

print(chemin)
# affichage graphique
# import matplotlib.pyplot as plt

# plt.figure(figsize=(10, 6))
# for trace in chemin:
#     plt.plot(trace, alpha=0.7)
# plt.xlabel('Lancer')
# plt.ylabel('Position')
# plt.title('Simulation de Mouvement Brownien')
# plt.show()
```
