```python
import numpy as np
def rejet(K, g, f, simul_Y):
    u = np.random.random()
    y = simul_Y()
    while True:
        if K*u*g(y) < f(y):
            return y
        else :
            u = np.random.random()
            y = simul_Y()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 10000
K = np.sqrt(2 * np.e / np.pi)

def sim_laplace():
    v = np.random.rand()
    return -np.sign(v - 0.5) * np.log(1 - 2 * abs(v - 0.5))

def f(x):
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-x**2 / 2)

def g(x):
    return 0.5 * np.exp(-abs(x))

samples = []
total_essais = 0

while len(samples) < N:
    Y = sim_laplace()
    U = np.random.rand()
    total_essais += 1
    if U <= f(Y) / (K * g(Y)):
        samples.append(Y)

samples = np.array(samples)

print(f"K = {K:.4f}")
print(f"Nombre moyen d'essais théorique = {K:.4f}")
print(f"Nombre moyen d'essais observé   = {total_essais / N:.4f}")
print(f"Moyenne = {samples.mean():.4f}, Écart-type = {samples.std():.4f}")

x = np.linspace(-4, 4, 300)
plt.figure(figsize=(8, 5))
plt.hist(samples, bins=60, density=True, alpha=0.6, color='steelblue', label="Échantillons simulés")
plt.plot(x, f(x), 'r-', linewidth=2, label="N(0,1) théorique")
plt.plot(x, K * g(x), 'g--', linewidth=1.5, label=f"K·g(x)")
plt.legend()
plt.tight_layout()
plt.show()
```
